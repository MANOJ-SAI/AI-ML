## **Lists:**

Lists are used to store multiple items in a single variable.

**List Items**
1. List items are **ordered, changeable, and allow duplicate values**.

### **Access List items**

```python


thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]

print(f"use the index to access the list : {thislist[1]}")
print(f"use the negative index to access the list : {thislist[-1]}") # return the last item of the list
print(f"use the range of indexs to access the list : {thislist[2:4]}") # return the list from start_index (included) to last_index - 1 (last_index is excluded)
```

### **Change List items**

```python
thislist = ["apple", "banana", "cherry"]
thislist[1] = "blackcurrant"
print(f"Change item in list using index:{thislist}")


thislist = ["apple", "banana", "cherry", "orange", "kiwi", "mango"]
thislist[1:3] = ["blackcurrant", "watermelon"]
print(f"Change a range of item values in list:{thislist}")
```

### **Add List items**

```python
thislist = ["apple", "banana", "cherry"]
thislist.insert(2, "watermelon")
print(f"insert the new item in list:{thislist}")


thislist = ["apple", "banana", "cherry"]
thislist.append("orange")
print(f"append the item to list1 at last:{thislist}")


thislist = ["apple", "banana", "cherry"]
tropical = ["mango", "pineapple", "papaya"]
thislist.extend(tropical)
print(f"append the list2 to list1 :{thislist}")


thislist = ["apple", "banana", "cherry"]
thistuple = ("kiwi", "orange")
thislist.extend(thistuple)
print(f"append the any iterable (list,tuple,set) to list1 :{thislist}")
```

### **Remove List Items**

```python
thislist = ["apple", "banana", "cherry", "banana", "kiwi"]
thislist.remove("banana")
print(f"remove the item in list using item:{thislist}") # it removes the first occurance


thislist = ["apple", "banana", "cherry"]
thislist.pop(1)
print(f"remove the item using index in list:{thislist}") #  the pop() method removes the last item when we didn't specify the index.


thislist = ["apple", "banana", "cherry"]
del thislist[0]
print(f"delete the item in list using index with del keyword:{thislist}")


thislist = ["apple", "banana", "cherry"]
del thislist # delete the entire list

thislist = ["apple", "banana", "cherry"]
thislist.clear()
print(f"empty the list:{thislist}")
```

### **Loop Through a List**

```python
thislist = ["apple", "banana", "cherry"]
print(f"iterate the list item using for: ",end=" ")
for x in thislist:
  print(x,end=" ")

print(f"\niterate the list item using range: ",end=" ")
thislist = ["apple", "banana", "cherry"]
for i in range(len(thislist)):
  print(thislist[i],end=" ")

thislist = ["apple", "banana", "cherry"]
i = 0
print(f"\niterate the list item using while: ",end=" ")
while i < len(thislist):
  print(thislist[i],end=" ")
  i = i + 1


fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = [x for x in fruits if "a" in x]   # syntax: newlist = [expression for item in iterable if condition == True]
print(f"\nusing the List Comprehension: {newlist}")
```

### **Join**

There are several ways to join two or more sets in Python.

1. The union() and update() methods joins all items from both sets.

2. The intersection() method keeps ONLY the duplicates.

3. The difference() method keeps the items from the first set that are not in the other set(s).

4. The symmetric_difference() method keeps all items EXCEPT the duplicates.

```python
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}
set3 = set1.union(set2)
print(f"union of set: {set3}")


set1 = {"a", "b", "c"}
set2 = {1, 2, 3}
set3 = set1 | set2
print(f"union of set using | operator: {set3}")


set1 = {"a", "b", "c"}
set2 = {1, 2, 3}
set3 = {"John", "Elena"}
set4 = {"apple", "bananas", "cherry"}
myset = set1.union(set2, set3, set4)
print(f"union of multiple sets: {myset}")


x = {"a", "b", "c"}
y = (1, 2, 3)
z = x.union(y)
print(f"union of set and tuple: {z}")


set1 = {"a", "b" , "c"}
set2 = {1, 2, 3}
set1.update(set2) # The update() method inserts all items from one set into another.
print(f"update the set1 with set2: {set1}") # The update() changes the original set, and does not return a new set.


set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}
set3 = set1.intersection(set2) # The intersection() method will return a new set, that only contains the items that are present in both sets.
print(f"intersection of set: {set3}")


set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}
set3 = set1 & set2 # The & operator only allows you to join sets with sets, and not with other data types
print(f"intersection of set using & operator: {set3}")


set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}
set1.intersection_update(set2)
print(f"update the set1 with intersection of set2: {set1}")


set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}
set3 = set1.difference(set2)
print(f"difference of set: {set3}")


set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}
set3 = set1 - set2
print(f"difference of set using - operator: {set3}")


set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}
set1.difference_update(set2)
print(f"update the set1 with difference of set2: {set1}")


set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}
set3 = set1.symmetric_difference(set2)
print(f"symmetric difference of set: {set3}")


set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}
set3 = set1 ^ set2
print(f"symmetric difference of set using ^ operator: {set3}")


set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}
set1.symmetric_difference_update(set2)
print(f"update the set1 with symmetric difference of set2: {set1}")
```

## **Lambda**

1. A lambda function can take any number of arguments, but can only have one expression
2. lambda arguments : expression

```python
x = lambda a : a + 10
print(f"simple example of lambda {x(5)}")


def myfunc(n):
  return lambda a : a * n
mydoubler = myfunc(2)
print(f"doubles the value using the lambda: {mydoubler(11)}")


numbers = [1, 2, 3, 4, 5]
doubled = list(map(lambda x: x * 2, numbers))
print(f"using lambda with map function:{doubled}")


numbers = [1, 2, 3, 4, 5, 6, 7, 8]
odd_numbers = list(filter(lambda x: x % 2 != 0, numbers))
print(f"using lambda with filter function:{odd_numbers}")


students = [("Emil", 25), ("Tobias", 22), ("Linus", 28)]
sorted_students = sorted(students, key=lambda x: x[1])
print(f"using lambda with sorted function:{sorted_students}")
```

## **Map**

1. The map() function executes a specified function for each item in an iterable. The item is sent to the function as a parameter.
2. map(function, iterables)

```python
def myfunc(a, b):
  return a + b
x = map(myfunc, ('apple', 'banana', 'cherry'), ('orange', 'lemon', 'pineapple'))
print(f"map function:{list(x)}")
```

## **Filter**

1. The filter() function returns an iterator where the items are filtered through a function to test if the item is accepted or not.

2. filter(function, iterable)

```python
ages = [5, 12, 17, 18, 24, 32]
def myFunc(x):
  if x < 18:
    return False
  else:
    return True
adults = filter(myFunc, ages)
print(f"filter function: ",end="")
for x in adults:
  print(x,end=" ")
```

# **OOP for AI: classes, objects, inheritance**

```python
class Model:
  def __init__(self, name):  # __init__ : Constructor (runs when object is created)
    self.name = name          # self : Refers to the current object
  def train(self):
    print(f"Training {self.name}")
```

```python
model1 = Model("Base Model")
model1.train()
```

```python
class AIModel:
    def __init__(self, name, accuracy=0.0):
        self.name = name
        self.accuracy = accuracy

    def train(self, epochs):
        self.accuracy += epochs * 0.01
        print(f"{self.name} trained for {epochs} epochs")

    def evaluate(self):
        print(f"Accuracy: {self.accuracy:.2f}")

model= AIModel("Base Model")
model.train(10)
model.evaluate()
```

```python
# Inheritance

class AIModel:
    def train(self):
        print("Training model")

    def predict(self, x):
        pass

class NeuralNetwork(AIModel):
    def __init__(self, layers):
        self.layers = layers

    def predict(self, x):
        return f"Prediction using {self.layers} layers"

model = NeuralNetwork(5)
model.train()
model.predict("image")
```

```python
# Polymorphism

class Model:
    def train(self):
        print("Training generic model")

class CNN(Model):
    def train(self):
        print("Training Convolutional Neural Network")

class RNN(Model):
    def train(self):
        print("Training Recurrent Neural Network")

models=[CNN(),RNN()]
for model in models:
  model.train()
```

```python
# super() in Inheritance

class Model:
    def train(self):
        print("Training generic model")

class Transformer(Model):
    def train(self):
        super().train()
        print("Applying attention mechanism")

model = Transformer()
model.train()
```

```python
# Encapsulation: Wrapping data and methods together and restricting direct access to the data.

class Model:
    def __init__(self):
        self._weights = "secret weights"

    def show_weights(self):
        print(self._weights)

model = Model()
model.show_weights()
```

```python
# Abstraction: Showing only essential features and hiding implementation details.

from abc import ABC, abstractmethod

class Model(ABC):
    @abstractmethod
    def train(self):
        pass

class CNN(Model):
    def train(self):
        print("Training CNN")

cnn = CNN()
cnn.train()
```

```python
# Composition (HAS-A Relationship): A class contains another class as a part of it to reuse functionality instead of inheriting it.

class Optimizer:
    def step(self):
        print("Updating weights")

class Model:
    def __init__(self):
        self.optimizer = Optimizer()

    def train(self):
        self.optimizer.step()

model = Model()
model.train()
```

# **File handling: CSV, JSON, text**

**File Modes:**
1. 'r'  -> Read
2. 'w'  -> Write (overwrites)
3. 'a'  -> **Append**
4. 'rb' -> Read binary
5. 'wb'	-> Write binary

**Common Mistakes:**

1. Not using **with**   -> **Solution:** use the "with" for file reading and writing operations
2. Loading huge files into memory -> **Solution:** For reading large dataset use the chunks
3. Hardcoding file paths -> **Solution:** use the os module for filepaths
4. Mixing CSV & JSON incorrectly -> **Solution:** based on the filetype use the corresponoding file type functions

```python
# File Paths (Important)

import os

print(os.getcwd())     # current directory
print(os.listdir())    # files in directory
```

## **Text File handling**

```python
# Writing the text file
file = open("notes.txt","w")
file.write("Learning AI\n")
file.write("ML cover under AI")
file.close()
```

```python
# Reading the text file
file = open("notes.txt","r")
print(file.read())
file.close()
```

```python
# Using "with" for reading and writing (Best practice)
with open("notes.txt","r") as file:
  print(file.read())
```

```python
# Reading Line by Line (large files)

with open("notes.txt","r") as file:
  for line in file:
    print(line)
```

```python
# Text Data for NLP (Advanced): Reading large text dataset

with open("notes.txt", "r", encoding="utf-8") as file:
    text = file.read().lower()

print(text[:100])
```

## **CSV Files (Used for ML)**

```python
# Writing csv file

import csv

with open("output.csv","w",newline="") as file:
  writer = csv.writer(file)
  writer.writerow(["age","score"])
  writer.writerow([25,80])
  writer.writerow([30,90])
```

```python
# Reading csv file

import csv

with open("output.csv","r") as file:
  reader = csv.reader(file)
  for row in reader:
    print(row)
```

```python
# Reading csv as Dictionary (Recommended)

import csv

with open("output.csv","r") as file:
  reader = csv.DictReader(file)
  for row in reader:
    print(row)
```

```python
# Reading CSV with Pandas (REAL AI Workflow)

import pandas

df = pandas.read_csv("output.csv")
print(df)
```
