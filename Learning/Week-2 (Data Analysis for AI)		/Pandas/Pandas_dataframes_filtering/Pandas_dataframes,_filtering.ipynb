```python
import pandas as pd
import csv
import random
from faker import Faker
import json
```

```python
class File_Writer:
  def __init__(self):
    self.fake = Faker()

  def generate_rows(self,num_rows):
    rows=[]
    for _ in range(num_rows):
      rows.append({
          "Name": self.fake.name(),
          "Age": random.randint(22,50),
          "Salary": random.choice(range(200000,2000000,100000)),
          "City": self.fake.city()
      })
    return rows

  def write_csv(self,file_name,num_rows):
    with open(file_name,'w',newline="") as csvfile:
      rows = self.generate_rows(num_rows)
      csvwriter = csv.DictWriter(csvfile,["Name","Age","Salary","City"])
      csvwriter.writeheader()
      csvwriter.writerows(rows)

  def write_json(self,file_name,num_rows):
    with open(file_name,'w') as jsonfile:
      rows = self.generate_rows(num_rows)
      json.dump(rows,jsonfile,indent=4)

file_writer = File_Writer()
file_writer.write_csv("data.csv",10)
file_writer.write_json("data.json",10)
```

### **ROW SELECTION**

**iloc:**

* Uses Python slicing rules
* End index is EXCLUDED

**loc:**
* Uses label-based slicing
* End label is INCLUDED

```python
# Syntax:
# df.iloc[] → Index-based selection
# df.iloc[row_index, column_index]
```

```python
# Uses integer positions (0,1,2…)
df.iloc[0]
```

```python
df.iloc[1:3] # Selects rows at index 1 and 2.
```

```python
df.iloc[0:2, 1:3] # Specific rows & columns
# 0:2 → ROWS (Select rows starting at 0 and STOP BEFORE 2)
# 1:3 → COLUMNS (Select columns starting at 1 and STOP BEFORE 3)
# Python slicing rule:
# Start is included, End is excluded
```

```python
# df.loc[] → Label-based selection
```

```python
# df.loc[row_label, column_name]
df.loc[0]
```

```python
df.loc[0, 'Salary']
# Select specific row & column
```

```python
df.loc[0:2, ['Name', 'City']]
# End label is INCLUDED. so it return the 0,1,2. in iloc it return 0,1 because the end label is excluded
```

```python
# | Feature     | `iloc`         | `loc`          |
# | ----------- | -------------- | -------------- |
# | Uses        | Index numbers  | Labels / names |
# | Rows        | Position-based | Label-based    |
# | Columns     | Position-based | Column names   |

# Column selection = choose features
# Row selection = choose samples

# iloc = Python slice = end excluded
# loc = label slice = end included
```

## **Filtering Data**
- Filtering = keep rows that satisfy a condition

### **Conditional Filtering:**

```python
data = {
    "Name": ["Amit", "Sara", "John", "Neha", "Raj"],
    "Age": [22, 28, 35, 40, 30],
    "Salary": [40000, 55000, 70000, 80000, 50000],
    "City": ["Bangalore", "Hyderabad", "Delhi", "Bangalore", "Hyderabad"]
}

df = pd.DataFrame(data)
```

```python
df[df['Age'] > 30] #From df, keep rows where age is greater than 30
```

```python
df[df['Salary'] >= 50000]
```

### **Multiple Conditions:**

```python
df[(df['Age'] > 25) & (df['Salary'] > 60000)]

# age > 25 AND salary > 60000
```

```python
df[(df['City'] == 'Bangalore') | (df['City'] == 'Hyderabad')]
```

```python
df['Salary'] > 60000
```

### **isin() (VERY CLEAN & READABLE)**

```python
df[(df['City'] == 'Bangalore') | (df['City'] == 'Hyderabad')]
# It is not readable instead of using multiple conditions we use the isin()
```

```python
# Instead of long OR conditions
df[df['City'].isin(['Bangalore', 'Hyderabad'])]
```

### **between() (Range filtering)**

```python
df[df['Age'].between(25, 40)] # between() is inclusive (25 and 40 included).
# Filtering =
# condition → True/False → keep True rows
```

## **Data Transformation (Apply Functions):**

* Data transformation = changing your data into the form you need.

```python
import pandas as pd

data = {
    "name": ["Amit", "Sara", "John", "Neha"],
    "age": [22, 28, 35, 40],
    "salary": [40000, 55000, 70000, 80000],
    "city": ["Bangalore", "Hyderabad", "Delhi", "Bangalore"]
}

df = pd.DataFrame(data)
df
```

```python
# Applies a function to each value in a Series or along rows/columns in a DataFrame
df['salary_lakh'] = df['salary'].apply(lambda x: x / 100000) # Function is applied to each element in that column.
df
```

```python
df['age_group'] = df['age'].apply(lambda x: 'Young' if x < 30 else 'Adult')
df
```

```python
df.apply(lambda row: row['age'] + row['salary'], axis=1) # function is applied row by row
```

```python
# Applies a function to every single value in a DataFrame.
df_numeric = df[['age', 'salary']]
df_numeric.applymap(lambda x: x * 1.1) #Recent versions of pandas deprecated applymap() to simplify the API.
```

```python
df_numeric = df_numeric.map(lambda x: x * 1.1) # Applies a function to a single Series (column)
df_numeric

# This does the same thing: multiplies every cell by 1.1
# Works on numeric and string columns depending on your function
```

```python
df['name'].map(lambda x: x.upper())
```

```python
# apply() → column or row → transform
# applymap() → every cell in DataFrame → transform
# map() → single column → transform

# Apply = flexible column/row
# Map = element-wise (Series or whole DF)
# Applymap = old element-wise DF
```

## **Aggregation & Statistics (Data Science CORE):**

```python
import pandas as pd

data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'department': ['IT', 'HR', 'IT', 'HR', 'Finance'],
    'city': ['NY', 'LA', 'NY', 'LA', 'NY'],
    'salary': [70000, 50000, 80000, 55000, 90000],
    'experience': [3, 2, 5, 4, 6]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.mean(numeric_only=True) # Average of numerical columns

# uses mean when data is Symmetrical( A distribution where values are evenly spread around the center and mean ≈ median.)
# No extreme values (outliers = values far away from the rest of the data)
```

```python
df.median(numeric_only=True) # Middle value (robust to outliers)
# Use median to fill missing values safely.
# If data has outliers or skew → mean lies
# Median tells the truth

# Why not use mean with outliers?
# Because mean is sensitive to extreme values and can misrepresent the typical value.
# Median is robust and provides a better central tendency when data is skewed or contains outliers or missing values.
```

```python
df['salary'].sum() # Total
```

```python
print(df.min(numeric_only=True))
print(df.max(numeric_only=True))
# Lowest & highest values
```

```python
df.count() # Number of non-null values
```

```python
df.describe() # All stats in one shot
```

```python
df.groupby('department').mean(numeric_only=True) # group by department
```

```python
df.groupby('city')['salary'].sum() # Which city pays more total salary?
```

```python
# What is the average salary & experience per department?
df.groupby('department').agg({'salary': 'mean', 'experience': 'mean'})
```

```python
# Instead of writing multiple lines mean(),min(),max(),count() and etc. we use multiple columns + multiple aggregations
# Group data → apply multiple statistics → get insights
df.groupby('department')['salary'].agg(['mean', 'min', 'max', 'count']) # One column, many aggregations
```

```python
import pandas as pd

data = {
    'employee': ['A', 'B', 'C', 'D', 'E', 'F'],
    'department': ['IT', 'IT', 'HR', 'HR', 'Finance', 'Finance'],
    'city': ['NY', 'LA', 'NY', 'LA', 'NY', 'LA'],
    'salary': [70000, 80000, 50000, 55000, 90000, 85000],
    'experience': [3, 5, 2, 4, 6, 5]
}

df = pd.DataFrame(data)
```

```python
df.groupby('department').agg({
    'salary': ['mean', 'min', 'max'],
    'experience': ['mean', 'max']
}) # Multiple columns + multiple aggregations
```

```python
df.groupby('department').agg(
    avg_salary=('salary', 'mean'),
    max_salary=('salary', 'max'),
    avg_exp=('experience', 'mean'),
    emp_count=('employee', 'count')
)
# Messy column names are bad for ML pipelines. so we rename the column like this, This is production-level Pandas.
```

```python
df.groupby(['department', 'city']).agg(
    avg_salary=('salary', 'mean'),
    total_salary=('salary', 'sum')
) # Average salary by department AND city
```

```python
df
```

```python
dept_features = df.groupby('department').agg(
    dept_avg_salary=('salary', 'mean'),
    dept_max_exp=('experience', 'max')
)

df = df.merge(dept_features, on='department')
df
```

```python
# | Function | Use When            |
# | -------- | ------------------- |
# | mean     | Normal distribution |
# | median   | Outliers exist      |
# | sum      | Totals matter       |
# | min/max  | Range & anomalies   |
# | count    | Missing values      |
# | describe | First look          |
# | groupby  | Compare categories  |

# EDA: EDA is the process of analyzing and summarizing data to understand its structure, detect patterns, identify outliers, and guide feature selection before building machine learning models.
# EDA (Exploratory Data Analysis)= getting to know your data before doing anything serious
# EDA means:
  # Checking what columns exist
  # Understanding typical values
  # Finding weird values
  # Spotting patterns
  # Detecting missing data & outliers

# EDA in Real Life:
  # df.head()
  # df.shape
  # df.info()
  # df.describe()
  # df.groupby('department')['salary'].mean()

  # Visual check:
    # df['salary'].hist()
    # df.boxplot(column='salary')
#


# Insights = useful conclusions you learn from data (what the data is telling you)
```

```python
df.groupby('city').agg(
    avg_salary = ('salary','mean'),
    total_salary = ('salary','sum'),
    max_experience = ('experience','max')
)
```

```python
df.groupby(['department','city']).agg(
    avg_salary = ('salary','mean'),
    emp_count = ('employee','count')
)
```

## **Renaming & Modifying Columns:**

```python
import pandas as pd

data = {
    'Employee Name': ['A', 'B', 'C', 'D', 'E', 'F'],
    'department': ['IT', 'IT', 'HR', 'HR', 'Finance', 'Finance'],
    'city': ['NY', 'LA', 'NY', 'LA', 'NY', 'LA'],
    'Annual Salary ($)': [70000, 80000, 50000, 55000, 90000, 85000],
    'Years of Experience': [3, 5, 2, 4, 6, 5]
}

df = pd.DataFrame(data)
```

```python
df.rename(columns={'salary': 'annual_salary'}) # Rename only one column
```

```python
df.rename(columns={'salary': 'annual_salary'}, inplace=True) # Rename permanently
```

```python
df = df.rename(columns={
    'Employee Name': 'employee',
    'Annual Salary ($)': 'salary',
    'Years of Experience': 'experience'
})
df
```

```python
df['salary']
```

```python
df
```

```python
df.drop(columns=['employee']) # drop the column
```

```python
df['experience'] = df['experience'].astype(int) # Converts column type to int.
df[['salary', 'experience']] = df[['salary', 'experience']].astype(int) # convert multiple columns type to int

df['salary'] = df['salary'].fillna(0).astype(int) # safe way to convert the column type because the column may have NaN, strings, decimals.
```

```python
# Columns with Too Many Missing Values
# | Missing % | Action  |
# | --------- | ------- |
# | < 5%      | Fill    |
# | 5–30%     | Careful |
# | > 40%     | Drop    |

# Is it just an ID? → Drop
# Is it free text with no plan to encode? → Drop
# Is it mostly missing? → Drop
# Does it reveal the target? → Drop
# Would I know this at prediction time? → Drop
```
