## **rand():**

* It follows uniform distribution. which means all elements have the equal probability(equal likely).
* It gives the values from the 0 to 1 because in the real world we convert everthing into master scale like min=0 and max=1 for  normalized values and it is easy for us to solve the complex  problems. and It's not giving the negative values because we don't  have the negative probability.

* **Master scale:** It is the concept of scale which in range of 0 to 1
* **Normalization:** It is the process of converting the real world numbers into the (0 to 1). ( It is the process of bringing the different measurement to a common scale)
* **Normalized value:** It is the output of normalization process. The value which is in master scale.

**Formula:** Normalization process = ( x - min ) / ( max - min )
**For example:** x = 20, min = 10 and max = 30
* Normalization process = ( 20 - 10 ) / ( 30 - 10 ) = 0.5
* Normalized value for 20 is 5. This is the way we calculate the normalized values.

**Syntax:**
 * 3D array= np.random.rand((depth or layers or pages),(rows),(columns))
 * 2D array= np.random.rand((rows),(columns))

## **randn():**

* It follows the Standard Normal Distribution or Z-Distribution. It is the special normal distribution with mean(center) of "0".
and standard deviation (which means Standard jump size away from center) is "1".
* In Normalization (rand) is about the boundaries ( 0 to 1).
* In Standardization (randn)  is about centering at "0".
* In randn(), the mean is "0" which means it is balanced. It ensures the positive number like +1.3 there is equal chance of getting
a negative value like -1.3.

The 3 Golden Rules ( The 68-95-99.7 Rule):
* It is famous part of the standard normal distribution. becuase the step size (Standard Deviation) is "1".
*     Rule-1 (The Inner Circle): 68.2% of all data falls between the -1 and +1
*     Rule-2 (The Outer Circle): 95.4% of all data falls between the -2 and +2
*     Rule-3 (The Boundary): 99.7% of all data falls in between the -3 and +3

In standard normal distribution the numbers on the bottom axis ( -3, -2, -1, 0, 1, 2, 3 ) are called Z-scores.
Z-scores tells about the "How many sandard deviations are we away from the average"

**For Example:**
* x = 20, data = [ 0, 10, 20, 30, 40 ]
1. Mean(μ) = (sum of all values) / (Total No .of values)
           = ( 0 + 10 + 20 + 30 + 40 ) / 5 = 20
2. Standard Deviation (σ) = ( Find the "gap" (difference) of each number from the mean (20), then square it.
    **σ** =
      * ( 0 - 20 )² = (-20)² = 400
      * ( 10 - 20)² = (-10)² = 100
      * ( 20 - 20)² = (0)²   = 0
      * ( 30 - 20)² = (10)²  = 100
      * ( 40 - 20)² = (20)²  = 400
   * Find the average of those squared gaps (called variance)
      
       variance = ( 400 + 100 + 0 + 100 + 400 ) / 5 = 200

       Take the square root to get the standard deviation (σ) =
          σ = √200 = 14.14
3. Calculate the Z-score for x = 20
            Z = ( x - μ ) / σ

            Z = ( 20 - 20) / 14.14 = 0

The Standarized value for 20 is 0

# **Matrix & Linear Algebra (CORE of ML & DL):**

```python
A @ B # matrix multiplication

# matrix multiplication:
  # A(m x n) x B(n x p) => C(m x p)
# Rule:
#   Inner dimensions must match (Without matching inner dimensions, dot product is impossible.)
#   Outer dimensions become the result
# Example: A = ([[1,2,3],
#                [4,5,6]])
#          B = ([[5,6],
#                [7,8],
#                [9,10]])
#    A x B = ([[1*5 + 2*7 + 3*9 , 1*6 + 2*8 + 3*10],
#              [4*5 + 5*7 + 6*9 , 4*6 + 5*8 + 6*10]])
#          = ([[46,52],
#              [109,124]]) #Rows of the first matrix × Columns of the second matrix
# Sample case:
#     (2 × 3) × (3 × 4)
#          ↑     ↑   ← match
#     Result → (2 × 4)
# Special cases:
#  Matrix x Vector:
#        (3 x 4) x (4 x 1) = (3 x 1)
#        (1 x 3) x (3 x 2) = (1 x 2)
```

```python
matrix_1 = np.arange(4).reshape(2,2)
matrix_1
```

```python
np.linalg.det(matrix_1) # returns the determinant of the matrix
# Formula: A = ([[a,b],
#               [c,d]])
#          det(A)= a*d - b*c
# Example: A= ([[0, 1],
#               [2, 3]])
#           det(A) = 0*3 - 1*2 = 0 - 2 = -2
```

```python
np.linalg.inv(matrix_1) # returns the inverse of matrix

# When does an inverse exist?
# A matrix has an inverse only if:
#  It is square (n x n)
#  Its determinant ≠ 0
# Inverse does not exist Such matrices are called singular matrices
# formula: A= ([[a,b],
#               [c,d]])
#          det(A) = ad - bc (determinant of matrix)
#          A_inv = (1/(ad-bc)) * [[d, -b],
#                                 [-c, a]]
# example: A=([[0, 1],
#              [2, 3]])
#        A_inv = (1/(0*3-1*2))*[[3, -1],
#                               [-2, 0]]
#        A_inv = [[-1.5,0.5],
#                 [1   ,  0]]
```

```python
np.linalg.eig(matrix_1)

# Eigenvalue = strength, Eigenvector = direction
# This intuition is also correct:
  # Eigenvector → direction
  # Eigenvalue → amount of stretching (importance)


# Step 1: Matching Rule (CRITICAL)
# --------------------------------
# NumPy returns:
# 1. Eigenvalues
# 2. Eigenvectors
#
# IMPORTANT:
# Eigenvalues match eigenvectors COLUMN-WISE (not row-wise)

# Example output:
# Eigenvalues: [-0.56, 3.56]
# Eigenvectors:
# Column 0      Column 1
# [-0.87,      -0.27]
# [ 0.49,      -0.96]

# Pairing:
# - Eigenvalue -0.56 ↔ Eigenvector [-0.87, 0.49]
# - Eigenvalue  3.56 ↔ Eigenvector [-0.27, -0.96]


# Step 2: What do these numbers mean?
# ----------------------------------
# Eigenvector → a special direction that does NOT change direction
#               when the matrix is applied.
#
# Eigenvalue  → how much the data is stretched or shrunk
#               along that direction.


# Step 3: Interpreting THIS matrix (not PCA)
# ------------------------------------------
# Matrix: [[0, 1],
#          [2, 3]]

# Direction 1:
# Eigenvector: [-0.27, -0.96]
# Eigenvalue : 3.56
# Meaning:
# - Strong stretching (3.56x)
# - This is the dominant direction of the transformation

# Direction 2:
# Eigenvector: [-0.87, 0.49]
# Eigenvalue : -0.56
# Meaning:
# - Weak stretching (0.56x)
# - Negative sign means flipping of direction


# - PCA uses eigenvalues of a COVARIANCE matrix, not a raw matrix


# One-line takeaway:
# ------------------
# np.linalg.eig() finds special directions of a matrix where vectors
# are only scaled (not rotated), and eigenvalues tell how strong
# that scaling is.


# The Eigenvalue (3.56) tells you "How much info" is in a direction.
# The Eigenvector ([-0.27, -0.96]) tells you "Which direction" that info is in.
# In Machine Learning, Bigger eigenvalue = more important direction of transformation
```

# **Missing Values Handling (NumPy level):**

```python
import numpy as np

# Marks of 5 students in a Math test
# One student was late (NaN), one box is empty
marks = np.array([40, 55, np.nan, 60, np.nan])

# 1. Checking for missing values
print("np.isnan:",np.isnan(marks))
# Output: [False, False, True, False, True]

# 2. The "Contamination" Problem
print("np.mean:",np.mean(marks))
# Output: nan (Useless!)

# 3. The "Smart" Solution (nanmean)
average_marks = np.nanmean(marks)
print(f"(np.nanmean) Average of present students: {average_marks}")
# Output: 51.66 (Calculates (40+55+60)/3)

# 4. Finding the Topper
topper = np.nanmax(marks)
print(f"(np.nanmax)Highest marks found: {topper}")
# Output: 60.0
```

```python
print(np.std(marks))
# Output: nan

# ✅ nanstd (The "Smart" way)
consistency = np.nanstd(marks)
print(f"The variation in marks: {consistency:.2f}")
# Output: 0.82 (Calculated using only 10, 12, 11)
```

# **Sorting & Ranking:**

```python
import numpy as np

# Marks of students: Rahul(85), Anjali(98), Vikram(72), Priya(85)
names = np.array(["Rahul", "Anjali", "Vikram", "Priya"])
marks = np.array([85, 98, 72, 85])

# 1. np.sort
print(np.sort(marks)) # It simply arranges the numbers in ascending order (lowest to highest).

# 2. np.argsort
ranks = np.argsort(marks) # It doesn't return the sorted numbers; it returns the Indices (the original positions) of the numbers in sorted order.
print(ranks)
# (Means: Index 2 is smallest, Index 1 is largest)

# Using argsort to sort names!
print(names[ranks])

# 3. np.unique
print(np.unique(marks)) # It removes all duplicates and returns a sorted list of distinct values.

# 4. np.unique with counts
values, counts = np.unique(marks, return_counts=True) # It gives you the unique values AND tells you how many times each value appeared.
print(dict(zip(values, counts)))
```

```python
import numpy as np

A = np.array([[1, 2],
              [3, 4]])
U, S, VT = np.linalg.svd(A) # Singular Value Decomposition: The SVD function decomposes a matrix into orthogonal input directions, stretch factors, and orthogonal output directions.
# U → matrix of left singular vectors
# S → array of singular values # Singular values tell you how much a matrix stretches space in special directions (the singular vectors)
# VT → transpose of right singular vectors

# Vector -> A vector is an ordered collection of numbers that represents a quantity with both magnitude and direction.
#
# | Type   | Example         | Meaning               |
# | ------ | --------------- | --------------------- |
# | Scalar |  5              | Just a number         |
# | Vector |  [2, 3]         | Direction + size      |
# | Matrix |  [[1,2],[3,4]]  | Collection of vectors |
#
# Singular vectors are special unit vectors that a matrix transforms only by stretching (or shrinking) and rotating without distortion.

# Right singular vectors:
  # These are the input directions
  # They tell you which directions to feed into the matrix
# Left singular vectors:
  # These are the output directions
  # They tell you where the matrix sends those special inputs

print("left singular vectors:",U)
print("singular values:",S)
print("right singular vectors:",VT)
```

## **SVD real world Example**

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data

# Load a sample image (grayscale)
img = data.camera()  # shape (512, 512)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.show()
```

```python
# Compute full SVD
U, S, VT = np.linalg.svd(img, full_matrices=False)
print("Top 10 singular values:", S[:10])

# Big numbers → important structures
# Small numbers → fine details / noise
```

```python
k = 50  # keep top 50 components

# Build diagonal Sigma
Sigma = np.diag(S[:k])

# Approximate image
img_approx = U[:, :k] @ Sigma @ VT[:k, :]

# U is the matrix of left singular vectors (output directions).
# U[:, :k] means take only the first k columns — the most important output directions.
# Shape: (rows of image, k)

# Sigma is a diagonal matrix of singular values you are keeping.
# Size: (k, k)
# Each singular value tells you how much to stretch along that direction.

# VT is the transpose of the right singular vectors (input directions).
# VT[:k, :] means take only the first k rows — the most important input directions.
# Shape: (k, columns of image)

# First, U[:, :k] @ Sigma → stretches the top k left vectors by singular values.
# Then, multiply by VT[:k, :] → rotates/stretches them back into the original coordinate system.

# Display compressed image
plt.imshow(img_approx, cmap='gray')
plt.title(f"Compressed Image with k={k}")
plt.show()

# Original image: 512 × 512 = 262,144 pixels
# Compressed image using 50 singular values: 50 × (512+512+50) ≈ 51,200 numbers → 80% reduction
```

## **Real world example of np.linalg.norm:**

```python
import numpy as np

# Rows = users, Columns = movies
ratings = np.array([
    [5, 4, 0, 0],  # User 1
    [4, 5, 0, 0],  # User 2
    [0, 0, 5, 4],  # User 3
    [0, 0, 4, 5],  # User 4
    [5, 4, 5, 4],  # User 5
])

print("Ratings matrix:\n", ratings)
```

```python
# Normalize each user's ratings to unit length
user_norms = np.linalg.norm(ratings, axis=1, keepdims=True)

# axis=1 → compute per row (per user)
# keepdims=True → keep the same number of dimensions so division works neatly

print(user_norms)
ratings_normalized = ratings / user_norms
print("\nNormalized ratings:\n", ratings_normalized)
```

```python
# Cosine similarity matrix
similarity = ratings_normalized @ ratings_normalized.T
print("\nUser similarity matrix:\n", np.round(similarity, 2))
```

```python
user_id = 0  # User 1
user_ratings = ratings[user_id]

# Find movies not rated
unrated = np.where(user_ratings == 0)[0]

# Predict ratings using weighted average of similar users
predicted_ratings = []
for movie in unrated:
    # Weighted sum of ratings by similarity
    sim_users = similarity[user_id]
    weighted_sum = np.sum(sim_users * ratings[:, movie])
    sim_sum = np.sum(sim_users)
    predicted = weighted_sum / sim_sum
    predicted_ratings.append(predicted)

# Show predictions
for i, movie in enumerate(unrated):
    print(f"Movie {movie+1} predicted rating: {predicted_ratings[i]:.2f}")
```

```python
arr = np.array([[10, 50, 30],
                [20, 40, 60]])

# Get indices of max element along axis 1
indices = np.argmax(arr, axis=1).reshape(-1, 1)
print(indices)

# Use take_along_axis to select max per row
result = np.take_along_axis(arr, indices, axis=1) # flexible, works with multi-dimensional arrays and fancy indices
print(result)
```

```python
import numpy as np

# 2D array with shape (2 rows, 3 columns)
arr = np.array([
    [10, 50, 30],
    [20, 40, 60]
])

# ---------------------------------------------------------
# np.argmax(axis=1)
# ---------------------------------------------------------
# axis=1 → operate row-wise
# Returns the column index of the maximum value for each row
#
# Row 0 → max(10, 50, 30) = 50 → column index 1
# Row 1 → max(20, 40, 60) = 60 → column index 2
#
indices = np.argmax(arr, axis=1).reshape(-1, 1)
print(indices)
# [[1]
#  [2]]

# ---------------------------------------------------------
# Why np.take() is NOT suitable here
# ---------------------------------------------------------
# np.take applies the SAME indices to EVERY row.
# It cannot handle different column indices per row.
#
# Example (wrong for this use case):
# np.take(arr, [1], axis=1)
# → takes column 1 from all rows

# ---------------------------------------------------------
# np.take_along_axis()
# ---------------------------------------------------------
# np.take_along_axis selects elements slice-by-slice
# along the given axis.
#
# Here:
# Row 0 → take column indices[0] = 1 → arr[0, 1] = 50
# Row 1 → take column indices[1] = 2 → arr[1, 2] = 60
#
# This allows DIFFERENT column indices for DIFFERENT rows.
#
result = np.take_along_axis(arr, indices, axis=1)
print(result)

# np.take()        → same indices applied to all slices
# take_along_axis  → different indices applied per slice
```
