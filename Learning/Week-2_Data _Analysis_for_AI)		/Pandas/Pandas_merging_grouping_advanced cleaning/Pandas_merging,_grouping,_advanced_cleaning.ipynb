## **MERGING DATA**

```python
import pandas as pd

users = pd.DataFrame({
    "user_id": [1, 2, 3,5],
    "age": [25, 30, 22,23],
    "country": ["US", "UK", "IN","EU"]
})
```

```python
purchases = pd.DataFrame({
    "user_id": [1, 2, 1, 3,4],
    "amount": [100, 200, 150, 50,110]
})
```

```python
merged = pd.merge(users, purchases, on="user_id") # INNER JOIN (default) (A intersection B : common values in both sets)
print(merged)
```

```python
pd.merge(users, purchases, on="user_id", how="left") # LEFT JOIN (A set values)
```

```python
pd.merge(users, purchases, on="user_id", how="right") # Right (B set values)
```

```python
pd.merge(users,purchases,on="user_id",how="outer") # OUTER JOIN (A union B)
```

```python
X = pd.DataFrame({
    "age": [25, 30, 22]
}, index=[1, 2, 3])  # user_id as index

country = pd.DataFrame({
    "country": ["US", "UK", "IN"]
}, index=[1, 2, 3])
X
```

```python
X = X.join(country) # index-based joining
print(X)
```

```python
transactions = pd.DataFrame({
    "customer_id": [1, 2, 1, 4],
    "amount": [100, 200, 150, 300]
})
```

```python
merged = pd.merge(
    users,
    transactions,
    left_on="user_id", # users table column name
    right_on="customer_id", # transactions table column name
    how="left"
)
print(merged)
```

```python
activity = pd.DataFrame({
    "uid": [1, 2, 2, 3],
    "sessions": [10, 5, 8, 2]
})
```

```python
df = pd.merge(
    merged,
    activity,
    left_on="user_id",
    right_on="uid",
    how="left"
)
df
```

```python
df = df.drop(columns=["customer_id", "uid"]) # Clean-up
df
```

```python
# Alternative: Cleaner pipeline instead of dropping
transactions = transactions.rename(columns={"customer_id": "user_id"})
activity = activity.rename(columns={"uid": "user_id"})
print("transactions:\n", transactions)
print("activity:\n", activity)
df = (
    users
    .merge(transactions, on="user_id", how="left")
    .merge(activity, on="user_id", how="left")
)
df
```

```python
# | Situation  | pandas approach            |
# | ---------- | -------------------------- |
# | 1-to-1     | `merge()`                  | Each row in Table A matches exactly one row in Table B.
# | 1-to-many  | `groupby  agg  merge`    | One row in Table A matches many rows in Table B.
# | Time exact | `merge(on=[id, time])`     | Data depends on both entity AND time.
# | Time fuzzy | `merge_asof()`             |
# | ML labels  | `aggregate  merge labels` | Labels are outcomes, not inputs.

# Check cardinality before merging (cardinality means how many times a value can appear in a column)
# customers["id"].is_unique
# orders["customer_id"].is_unique

# Validate merge (super useful!)
# customers.merge(
#     profiles,
#     on="id",
#     validate="one_to_one"
# ) # "one_to_many","many_to_one"

```

## **GROUPING DATA (groupby):**
groupby lets you convert raw data into meaningful features by aggregating patterns across groups.

**Split  Apply  Combine**

- Split data into groups based on a key

- Apply a function to each group

- Combine the results back

```python
import pandas as pd

df = pd.DataFrame({
    "name": ["A", "B", "C", "D", "E", "F"],
    "department": ["IT", "IT", "HR", "HR", "Sales", "Sales"],
    "city": ["NY", "LA", "NY", "LA", "NY", "LA"],
    "team": ["A", "A", "B", "B", "C", "C"],
    "salary": [60000, 55000, 50000, 52000, 70000, 72000],
    "age": [25, 28, 35, 32, 40, 38]
})
```

```python
# 1. Basic Grouping
#  groupby returns a GroupBy object, not a DataFrame.
df.groupby("department")
```

```python
# 2. Aggregations
df.groupby("department")[["salary","age"]].mean()
# if I don't use salary then it try to calulcate mean non numeric columns also it leads to error
```

```python
df.groupby("city")["salary"].sum()
# grouby sum
```

```python
df.groupby("team").count()
# groupby count: count rows per group
# Counts non-null values per column.
# If there are NaNs  count()  number of rows.
```

```python
df.groupby("team").size()
# Counts rows in each group, no matter whats inside (counts Nan's also)
# Better for row count
```

```python
# 3. Multiple Aggregations;
df.groupby("department").agg({
    "salary": ["mean", "max"],
    "age": "mean"
})
```

```python
df.groupby("department").agg(
    avg_salary=("salary", "mean"),
    max_salary=("salary", "max"),
    avg_age=("age", "mean")
)
# adding column names
```

```python
# 4. Groupby + Filtering
df.groupby("department").filter(
    lambda x: x["salary"].mean() > 50000
)
```

```python
# 1. groupby().mean() vs agg()
  # mean()  quick
  # agg()  flexible & scalable

# 2. count() vs size()
  # count()  ignores NaNs
  # size()  counts rows

# 3. Filtering rows vs groups
  # Row filtering  df[df.col > x]
  # Group filtering  groupby().filter()
```

```python
# Feature engineering is the process of turning raw data into useful inputs (features) for a machine-learning model.
# Better features matter more than better models.
# Feature Engineering example:
user_features = (
    df.groupby("department")
      .agg(
          avg_salary=("salary", "mean"),
          salary_std=("salary", "std"),
          headcount=("name", "count")
      )
)
user_features
```

```python
# KPI (Key Performance Indicator):A KPI is a metric that measures how well a business or system is performing against a goal.
# | Feature Engineering     | KPI                             |
# | ----------------------- | ------------------------------- |
# | Used for ML models      | Used for monitoring & decisions |
# | Input to model          | Output for humans               |
# | Row-level or aggregated | Usually aggregated              |
# | Predictive              | Descriptive                     |


# KPIs (for business):
  # Monthly revenue
  # Avg order value
  # Active users

# Features (for ML churn model):
  # Avg spend last 30 days
  # Number of transactions
  # Days since last purchase

# KPIs describe how the business is doing; features help the model learn patterns.
```

## **ADVANCED DATA CLEANING**

- Data cleaning ensures consistency, correctness, and numeric compatibility before feature engineering and modeling.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "id": [1, 2, 2, 3],
    "name": [" Alice ", "BOB", "bob", None],
    "salary": [50000, np.nan, 60000, 5000000],
    "age": ["25", "30", None, "40"],
    "date": ["2024-01-01", "2024/01/02", None, "2024-01-04"]
})
df
```

```python
# 1. Detect missing values:
df.isnull()
# Use this to locate missing data.
```

```python
df.isnull().sum()
# Counts missing values per column.
```

```python
df
```

```python
# 2. Advanced fillna()
df.ffill() # Forward fill (ffill) (old version : df.fillna(method="ffill"))
# Fills missing value using previous rows value.
```

```python
df.bfill() # Backward fill (bfill) (old version : df.fillna(method="bfill"))
# Fills missing value using next rows value.
```

```python
df["salary"].bfill() # column specific
```

```python
df["salary"].fillna(df["salary"].mean())
# df.fillna(df.mean()) works only for numeric columns.
```

```python
# 3. Remove Duplicates
df.duplicated()
# Returns True for duplicated rows.
```

```python
df
```

```python
df.drop_duplicates()
# Remove full duplicates
```

```python
df.drop_duplicates(subset=["id"])
# Remove duplicates based on specific columns
# Keeps first occurrence by default
```

```python
df.drop_duplicates(subset=["id"], keep="last")
```

```python
# 4. Fix Data Types
df["age"] = df["age"].astype(float)
# Convert column type
# This will FAIL if there are non-numeric strings.
df
```

```python
df["age"] = pd.to_numeric(df["age"], errors="coerce")
# safer way to covert column type
df
```

```python
df["date"] = pd.to_datetime(df["date"], errors="coerce")
# Convert date column
# If 'raise', then invalid parsing will raise an exception.
# If 'coerce', then invalid parsing will be set as NaT.
# If 'ignore', then invalid parsing will return the input.
df
```

```python
# 5. String Cleaning
df["name"] = df["name"].str.lower()
# Lowercase
df
```

```python
df["name"] = df["name"].str.strip()
# Remove leading/trailing spaces
df
```

```python
df["name"] = df["name"].str.replace(" ", "", regex=False)
# Remove spaces inside strings
df
```

```python
# 6. Handling Outliers

df = df[df["salary"] < 1_000_000]
# Simple threshold filtering : Removes unrealistic values.
df
```

```python
df = (
    df
    .drop_duplicates()
    .assign(
        name=lambda x: x["name"].str.lower().str.strip(),
        age=lambda x: pd.to_numeric(x["age"], errors="coerce"),
        date=lambda x: pd.to_datetime(x["date"], errors="coerce")
    )
)

df["salary"] = df["salary"].fillna(df["salary"].median())
df
```

## **COLUMN ENGINEERING**
- Column engineering (feature preparation) is creating new columns from existing data to help ML models or analysis.

```python
df
```

```python
# 1. Scaling / Normalization
df['new_col'] = df['salary'] / 100000
df
```

```python
# 2. Creating Boolean Feature
df['is_high_salary'] = df['salary'] > 7000
df
```

```python
# Converts numeric thresholds into categorical signals

# Useful for:
  # ML models that handle categorical data well
  # Feature engineering for insights/KPIs
  # Can also convert to 0/1 for ML:
  # df['is_high_salary'] = (df['salary'] > 70000).astype(int)

# Column engineering = creating new columns/features
# Scaling (salary / 100000)  makes numbers easier for models
# Boolean/Flag (salary > 70000)  highlights important thresholds
# Used for ML models, feature engineering, and KPIs
```

```python
# Key Idea in Feature Engineering
# | Technique                | Example                                                   |
# | ------------------------ | --------------------------------------------------------- |
# | Transformation / Scaling | salary / 100000                                           |
# | Boolean feature / Flag   | is_high_salary = salary > 70000                           |
# | Aggregation              | avg_salary_per_dept = df.groupby("dept")["salary"].mean() |
# | Interaction              | new_col = salary * years_experience                       |

# These features make models smarter and can capture patterns that raw data cannot.
```
